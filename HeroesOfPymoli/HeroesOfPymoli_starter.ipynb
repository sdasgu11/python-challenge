## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Final Total count of unique students.
unique_person = purchase_data["SN"].nunique() 
print(unique_person)

# count the number of male players
purchase_data.head()
male_players_df = purchase_data.loc[(purchase_data["Gender"] == "Male")]
male_count = male_players_df["SN"].nunique() 
print(male_count)
print((male_count/unique_person)*100)

# count the number of female players
female_players_df = purchase_data.loc[(purchase_data["Gender"] == "Female")]
female_count = female_players_df["SN"].nunique() 
print(female_count)

# count the number of other players
other_players_df = purchase_data.loc[((purchase_data["Gender"] != "Female") & (purchase_data["Gender"] != "Male"))]
other_count = other_players_df["SN"].nunique() 
print(other_count)



#Percentage and Count of Other / Non-Disclosed
non_disclosed_count = unique_person-(male_count+female_count)
print(non_disclosed_count)
print('{0:.2f}'.format(non_disclosed_count/ unique_person * 100))

mgender_demo_df = pd.DataFrame({"Gender":["Male"],"Total Count":[male_count],"Total Percentage":['{0:.2f}'.format(male_count/unique_person*100)]})
fgender_demo_df = pd.DataFrame({"Gender":["Female"],"Total Count":[female_count],"Total Percentage":['{0:.2f}'.format(female_count/unique_person*100)]})
ogender_demo_df = pd.DataFrame({"Gender":["Others"],"Total Count":[non_disclosed_count],"Total Percentage":['{0:.2f}'.format(non_disclosed_count/ unique_person * 100)]})

concat_df = pd.concat([mgender_demo_df,fgender_demo_df,ogender_demo_df])

#concat_df.reset_index().head()
concat_df.reset_index(drop=True).head()
#concat_df.head()
                               
                               

```
